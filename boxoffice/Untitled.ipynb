```python
import pymongo
import pandas as pd
client = pymongo.MongoClient("mongo")
database = client['boxoffice']
collection_box = database['scrapy_items']

releases_all = collection_box.distinct("releaseID")
release_arg = ",".join(releases_all)
```

```python
from boxoffice.spiders.imdb import ImdbSpider
import scrapy
from scrapy.crawler import CrawlerProcess

if __name__ == "__main__":
    process = CrawlerProcess()
    process.crawl(ImdbSpider,releases_args = release_arg)
    process.start()
```
